# Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5, 4))
plt.title('Reservation status count')
bar_plot = plt.bar(['Not canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)

# Display values on top of the bars
for bar in bar_plot:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Add display values on top of the bars
for p in ax1.patches:
    height = p.get_height()
    if not np.isnan(height):
        ax1.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', xytext=(0, 5),
                     textcoords='offset points', fontsize=8)

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status in different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not canceled', 'Canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel.head(20)
```

```python
city_hotel.head(20)
```

```python
plt.figure(figsize = (25,10))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)

plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')

plt.legend(fontsize = 20)
plt.show()
```

```python
#extracting the month component from the 'reservation_status_date' column using the dt.month
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(20, 10))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Add display values on top of the bars
for p in ax1.patches:
    height = p.get_height()
    if not np.isnan(height):
        ax1.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='bottom', xytext=(0, 5),
                     textcoords='offset points', fontsize=8)

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status per month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['Not canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR for each month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel('Month', fontsize=16)  
plt.ylabel('Average Daily Rate (ADR)', fontsize=16)
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]

# Define custom colors
colors = ['teal', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'gray', 'brown', 'red']

plt.figure(figsize=(8, 8))
plt.title('Top 10 countries with most reservation canceled')
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index, colors=colors)
plt.show()
```
